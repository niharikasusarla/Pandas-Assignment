```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/sales_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df
```

```python
city_sales = df.groupby('City')['Total Sales'].sum().reset_index()
```

```python
city_sales = city_sales.rename(columns={'Total Sales': 'Total Sales per City'})
print("\nTotal sales per city:")
print(city_sales)
```

```python
product_sales = df.groupby('Product')['Total Sales'].sum()
```

```python
highest_sales_product = product_sales.idxmax()
print( highest_sales_product)
```
